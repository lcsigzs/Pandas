```python
df.groupby(["IdCustomer"]).agg({"Points": 'sum', "UUID": "count", "DtTransaction": 'max'}).rename(columns={"Points":"Valor", "UUID":"Frequencia","DtTransaction":"UltimaData"}).reset_index()
```

```python
def recencia(x):
    diff = datetime.datetime.now() - x.max()
    return diff.days
```

```python
df.groupby(["IdCustomer"]).agg({"Points": 'sum', "UUID": "count", "DtTransaction": ['max', recencia]}).rename(columns={"Points":"Valor", "UUID":"Frequencia","DtTransaction":"UltimaData"}).reset_index()
```
