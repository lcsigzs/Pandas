```python
import pandas as pd
```

```python
data_users = {
    "id": [1,2,3,4],
    "nome":["Teo","Mat","Nah","Mah"],
    "idade":[31, 31, 32, 32]
}
```

```python
df_user = pd.DataFrame(data_users)
df_user
```

```python
data_transacoes = {
    "id_user": [1, 1, 1, 2, 3, 3],
    "vl": [432, 532, 123, 6, 4, 87],
    "qtProdutos": [2, 1, 3, 6, 10, 2]
}
```

```python
df_transacao = pd.DataFrame(data_transacoes)
df_transacao
```

```python
#mesclando 2 dataframes
#merge usa-se no df que ira ficar na esquerda, depois passa o da direita como parametro, how para saber qual o shape voce vai mante->
#e depois qual argumento dos 2 voce vai usar para "comparar"
df_transacao.merge(df_user, how="left", left_on="id_user", right_on="id")
```
