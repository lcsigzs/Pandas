```python
import pandas as pd
```

```python
data_01 = {
    "id": [1,2,3,4],
    "nome":["Teo","Mat","Nah","Mah"],
    "idade":[31, 31, 32, 32]
}
```

```python
df_1 = pd.DataFrame(data_01)
df_1
```

```python
data_02 = {
    "id": [5,6,7,8],
    "nome":["Jose","Natan","Arnaldo","Mario"],
    "idade":[23, 33, 19, 21],
}
```

```python
df_2 = pd.DataFrame(data_02)
df_2
```

```python
pd.concat([df_1,df_2]).reset_index(drop=True)
```

```python
data_03 = {
    "sobrenome":["Calvo","Silva","Costa","Souza"],
    "renda":[3100, 3100, 3200, 3200]
}
```

```python
df_3 = pd.DataFrame(data_03).sort_values(['renda','sobrenome'], ascending=(False, True))
df_3
```

```python
pd.concat([df_1, df_3], axis=1)
```
