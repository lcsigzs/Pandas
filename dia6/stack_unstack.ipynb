```python
import pandas as pd
```

```python
df = pd.read_csv("../data/bia_consolidado.csv", sep=';')
df
```

```python
df_stack = df.stack().reset_index().rename(columns={"level_3":"Tipo de Homicido", 0:"Qtde"})
```

```python
df_unstack = (df_stack.set_index(['cod','nome','período', 'Tipo de Homicido'])
                      .unstack()
                      .reset_index())
```

```python
homicidios = df_unstack['Qtde'].columns.tolist()
indentificadores = df_unstack.columns.droplevel(1).tolist()[:3]

df_unstack.columns = indentificadores + homicidios
df_unstack
```
