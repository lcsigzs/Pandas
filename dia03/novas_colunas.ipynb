```python
data = {
    "nome": ["Teo", "nah", "Maria", "Lara"],
    "recencia" : [1, 30, 10, 45],
    "valor" :[100, 2000, 15, 500],
    "frequencia":[2, 5, 1, 15]
}
```

```python
df_crm = pd.DataFrame(data)
```

```python
def rfv(row):
    nota = 0

    if row['recencia'] <= 10:
        nota += 10
    elif 10 < row['recencia'] <= 30:
        nota += 5
    elif row['recencia'] > 30:
        nota += 0

    if row['valor'] > 1000:
        nota += 10
    elif 100 <= row['valor'] < 1000:
        nota += 5
    elif row['valor'] < 100:
        nota += 0

    if row['frequencia'] > 10:
        nota += 10
    elif 5 <= row['frequencia'] < 10:
        nota += 5
    elif row['frequencia'] < 5:
        nota += 0

    return nota
```

```python
rfv(1, 100, 2)
```

```python
df_crm.apply(rfv, axis=1)
```
