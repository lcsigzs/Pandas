```python
import pandas as pd
```

```python
df = pd.read_csv("../data/customers.csv", sep=";")
df
```

```python
# ordenando pela pontuação
df.sort_values(by="Points", ascending=False, inplace=True)
df.rename(columns={'Name': 'Nome', 'Points': 'Pontos'}, inplace=True)
df
```

```python
df.sort_values(by=["Pontos", 'Nome'], ascending=[False, True])
df
```

```python
df = (df.sort_values(by=["Pontos", 'Nome'],
                     ascending=[False, True])
      .rename(columns={'Name': 'Nome', 'Points': 'Pontos'}))
df
```
